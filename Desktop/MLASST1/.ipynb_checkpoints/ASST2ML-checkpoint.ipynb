### 1 Introduction
The most critical component of electrical power systems is transmission lines, owing to their ability to transmit electric power from generation plants to distribution networks. Similarly, another crucial concern in achieving efficient delivery of electric power is the occurrence of electrical faults in the transmission lines. These faults can affect the stability and reliability of the electrical distribution, potentially causing safety hazards and financial losses. Various types of faults may occur in transmission lines, including elevated temperatures that increase the conductor resistance and create conditions for insulation breakdown, leading to short circuits and other faults. An increase in temperature can also result in an increased resistance, which may reduce the current flowing through the conductor, leading to greater power losses. Higher temperatures can also decrease the resistance of certain insulating materials, increasing the current leakage and the likelihood of fault conditions. When a fault occurs, the robustness of the electricity distribution equipment for power system protection serves an essential function in isolating the area in which the fault is identified. Timely detection of these faults is crucial for proper intervention and mitigation of larger systemic failures. A reliable fault detection system is necessary to provide a robust, time-efficient, and secure means of resilient operation. The objective of this project is to compare two machine learning models, Support Vector Machines (SVM) and Random Forests (RF), for binary classification to gain insights into the strengths and weaknesses of a complex dynamic system.

### 1. Support Vector Machines 
**Support Vector Machines (SVMs)** are considered a supervised learning model that can be applied to both classification and regression problems. SVM classifier was first introduced in the 1990s by Vladimir Vapnik and has been implemented across various domains, increasing its popularity due to its theoretical foundation [2].

**1.1 Explain the main concepts behind the algorithm**

SVM's main concept is to find the optimal hyperplane that separates the data points into their respective classes in the most efficient way. The hyperplane is determined by support vectors(i.e., the data points that are closest to the decision boundary). Overall, it is considered an optimization problem to find the maximum margin that separates parts of the data from each other[2][3].

- #### Important Terms 
- **Decision Boundary** : The optimal hyperplane that separates the data points into respective classes. 
- **Hyperplanes** : A hyperplane is determined by the respective support vectors, it identifies the distance from the decision boundary. 
- **Support Vectors** : The data points that are laid on the margin of the hyperplane, meaning they are closest to the decision boundary. 
- **Margin/ Distance** : The distance between the decision boundary and the support vector datapoints.

- #### The Kernel Trick
There is a limited amount of linear classifier functions, and thus this leads to a likelihood of under-fit data. Yet, it is possible to have both a linear model as well as a set of nonlinear functions. This is done via the Kernel trick (with maximum margin hyperplanes). By the use of the kernel trick for SVMs, the maximum margin hyperplane is now fit in a feature space $F$.  $F$ is a non-linear map $\Phi: R^N \rightarrow F$ from the original input space to a higher dimensional space [4]. This suggests that the SVM algorithm can handle nonlinear problems by the application of kernel functions that map the data into a higher dimensional space  making it linearly separable. Kernel functions include polynomial and radial basis functions [4]. 

As can be seen in Figure 3, the idea of kernel SVM is to map the training data into a higher dimensional feature space via $\Phi$ and construct a separating hyperplane with maximum margin, overall yielding a nonlinear decision boundary in input space [4].
$$
\begin{align}
\Phi: \mathbf{R}^2 \rightarrow \mathbf{R}^3,\left(x_1, x_2\right) \rightarrow\left(z_1, z_2, z_3\right) \equiv\left(x_1^2, \sqrt{2} x_1 x_2, x_2^2\right).
\end{align}
$$

- **1.2 Discuss how the algorithm works in a straightforward manner**

The fundamental objective of SVM is to identify and construct an optimal hyperplane that efficiently separates the differing classes of data yet still maximizes the margin between them. For example, in a binary classification problem, the SVM's main objective is to find the decision boundary that efficiently separates the two classes. This is done by solving a quadratic programming problem that can maximize the distance between the decision hyperplane to the nearest data points from each class [2].  


The decision boundary represents the equation of a line in compact vector form:

$$
\begin{align}
(\vec{w} \cdot \vec{x})+b=0 \quad \mathbf{w} \in \mathbf{R}^N, b \in \mathbf{R} 
\end{align}
$$

**Where** : 
$b$: the intercept of the line, ${w}$: the coefficients, ${x}$: the input datapoints.

According to [2][4], by assuming that the training sample is separable, the rescaling of ${w}$ and $b$ can be done such that the points closest to the hyperplane can satisfy 
$$
\begin{align}(\vec{w} \cdot \vec{x})+b= 0 \end{align}
$$ 


The equations for the points on the margin for two classes (i.e., the Equation for support vectors are: 


$$
\begin{gather}
(\vec{w} \cdot \vec{x}) + b = 1 \\
(\vec{w} \cdot \vec{x}) + b = -1
\end{gather}
$$ 




<div align="center">
    <img src="F333.png" alt="Fundamental principle of SVM" width="50%"><br>
    <strong>Fig 4: Fundamental principle of SVM</strong>
</div>



We solve for $b$ and $w$, by finding the decision boundary that leads to maximizing the margin between the two classes. The distance is calculated by the normal/ orthogonal vector(e.g., vector $w$ that is perpendicular to the decision boundary line) [2][4]. The distance is given by:
$$
\begin{align}
\frac{1}{\|\vec{w}\|}
\end{align}
$$

The size of the margin is twice that distance:
$$
\begin{align}
\frac{2}{\|\vec{w}\|}
\end{align}
$$

Overall, this suggests that the optimal hyperplane is unique in minimizing $(w • w)$ under the constraints. By constructing an optimal hyperplane,  Quadratic programming as a type of mathematical optimization problem will be applied where the objective function is quadratic and the constraints are linear. This all falls under the category of convex optimization [2][4].

**Maximizing the margin**: We maximizing $(\frac{2}{\|\mathbf{w}\|})$ is equivalent to minimizing $(\|\mathbf{w}\|)$ or $\frac{1}{2}\|\mathbf{w}\|^2$

**Constraint**: subject to

$$(\mathbf{w} \cdot \mathbf{x} + b) \geq 1, \quad \forall \mathbf{x} \text{ of class 1}$$
$$(\mathbf{w} \cdot \mathbf{x} + b) \leq -1, \quad \forall \mathbf{x} \text{ of class 2}$$

### Optimization Problem Formulation:
$$
\begin{align}
\left(\mathbf{x}_n, y_n\right) &\in \mathbf{R}^N, \quad \mathbf{x} \in \mathbf{R}^N \\[0.5cm]
\text{Minimize} \quad &\frac{1}{2}\|\mathbf{w}\|^2 \\[0.5cm]
\text{subject to} \quad & y_i \left( \vec{w} \cdot \vec{x}_i - b \right) \geq 1, \; \forall i, \quad y \in \{-1, +1\}
\end{align}
$$.

#### 2. Random Forest

Random Decision Forests, firstly introduced 1995 by a computer scientist in IBM Tin Kam Ho, who was able to achieve this discovery by elaborating on the foundations placed by Leo Brierman, who initially introduced and named it the Random Forests algorithm back in 2001. It is essential a supervised algorithm that is based on a collection of learning techniques that combine numerous decision trees to create a robust learning model [5][6]. 

- **2.1 Explain the main concepts behind the algorithm**
  
The algorithm, Random Forests combines multiple decision tree models. Each tree is then trained via a differing set of data, and thus the predictions are then complied to make the final prediction. 


- #### Important Terms

- **Tree**: Consists of three basic features – branches, nodes, and leaves. 
- **Nodes**: An element of a tree, the point where point(s) branches come out. 
- **Root node**: Initial node of the tree, the origin of the structure
- **Decision  node**: The predecessor of another node. 
-  **Leaf node**: External node, meaning where no branches originate.
-  **Bagging (Bootstrap Aggregation)**: A technique that pertains to creating numerous training sets via the process of drawing random samples, acting as replacements from the original data. Each of the bootstrap samples is then used to train a separate decision tree. 
-  **Entropy**: The entropy measures the impurity of the data gathered at the node- it quantifies uncertainty in the data at each node. 
-  **Information Gain**: It is the difference between the original Entropy and the weighted average Entropy of the subsets created after the split.

- **2.2 Discuss how the algorithm works in a straightforward manner**

Without relying on one tree, Random Forest constructs multiple trees, every tree is trained on a slightly different set of data. The process of how decision is achieved in the prediction process Decision tree algorithm is as follows: 
it is based on a sequence of comparisons between the samples’ input and the  pre-defined  threshold values. This process of comparisons leads the model to reach the optimal prediction at the leaf node. To obtain the optimal split, the algorithm can consider the split that maximizes the information gain (i.e., it minimizes the impurity), overall leading to more homogenous nodes. Thus, it is the measure of the feature's effectiveness, such as expecting the reduction in entropy, in overall classifying the training data[8].

The information gain, Gain (S, A) of a feature A, relative to a collection of training examples S, is defined as:
$$
\begin{align}
\operatorname{Gain}(S, A) \equiv \operatorname{Entropy}(S)-\sum_{v \in V \text { values }(A)} \frac{\left|S_v\right|}{|S|} \operatorname{Entropy}\left(S_v\right)
\end{align}
$$

Where:

- $S$: a collection of training examples.

- $\text{Values}(A)$: the set of all possible values for feature $A$.

The Entropy can be calculated as follows:
$$
\begin{align}
\operatorname{Entropy}(S) \equiv \sum_{i=1}^c-p_i \log _2 p_i
\end{align}
$$

Where:
$p_i$ is the proportion of S belonging to class i.
  
The higher the Information Gain, the better the split. Overall, the algorithm expects the split that minimize entropy in the resulting branches, thus leading to more robust predictions.).
)).




It can be seen in Figure 6, the algorithm, Random Forest, is based on a collection of decision tree. The algorithm gains the collective of multiple decision trees to obtain more robust, effective and accurate predictions by Bootstrap Aggregation where each tree is trained separately on a differing subset of the original data.

### 3. Dataset Overview

The dataset utilized in this study is obtained from [10] and is publicly available on Kaggle, allowing the research findings reproducibility. In this dataset, MATLAB Simulink was implemented to design power system simulation for power components at the three stages: generation, transmission, and distribution, with transmission lines under normal conditions as well as under the main types of fault conditions. The voltage-current data were generated and collected with measured line voltages and line currents at the output side of the power system. The dataset contains both Binary and Multiclass Classification.

- **3.1 Classification Types**
  
 - ##### Binary Classification

Approximately **12,000** data observations were collected and subsequently labeled into two distinct classes: **0**, representing normal operating conditions (i.e., no fault), and **1**, indicating an abnormal condition (i.e., faulty line).

 - ##### Multiclass Classification

For the multiclass classification task, around **8,000** data observations were gathered to represent various fault types, categorized into five distinct classes. The classifications are as follows:

- **[0, 0, 0, 0]**: No Fault
- **[1, 0, 0, 1]**: Line-to-Ground (LG) Fault
- **[0, 0, 1, 1]**: Line-to-Line (LL) Fault
- **[1, 0, 1, 1]**: Line-to-Line-to-Ground (LLG) Fault
- **[0, 1, 1, 1]**: Line-to-Line-to-Line (LLL) Fault (affecting all three phases)
- **[1, 1, 1, 1]**: Line-to-Line-to-Line-to-Ground (LLLG) Fault (three-phase symmetrical fault)

```python
#  Select the input features(6 input features) 
X = df[['Ia', 'Ib', 'Ic', 'Va', 'Vb', 'Vc']]
print(X)
```

```python
# A descriptive statistic of the input feature 
X.describe()
```

```python
#select the target/response variable
y= df[['Output (S)']]
print(y)
```

```python
# Split the dataset into training and testing sets

X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.25, random_state=42)
```

```python
print(X_train.shape)  
print(X_test.shape)   
print(y_train.shape)  
print(y_test.shape)   
```

```python
# Scaling the training dataset with min-max scaler
min_max_scaler= preprocessing.MinMaxScaler()
X_train_scaled= min_max_scaler.fit_transform(X_train)
X_test_scaled= min_max_scaler.fit_transform(X_test)
print(X_train_scaled)
```

```python
print(y_train)
```

#### 5. Model Design and Training
- **5.1 SVM  Design and Training**

```python
# define the model
classifier1= SVC()  

# Hyperparameter tuning for the SVM model using GridSearchCV
# C: regularization parameter:The strength of the regularization is inversely proportional to C(i.e., lower c stronger the reqularization)
param_grid1 = [{ 'C': [0.1, 0.5, 1],  'kernel': ['linear']}, # kerenal funcation
               { 'C': [0.1, 0.5, 1],  'kernel': ['rbf'], 'gamma':[0.1, 0.3, 0.9]}] # kernel coefficient
  

# Do a grid search 
grid1 = GridSearchCV(estimator= classifier1, param_grid=param_grid1, scoring = 'accuracy', cv= 5 , n_jobs=-1)

# grid search start
start_time = time.time()

# Fit the model- Train the SVM  Classifier with X_train & y_train, the kernel used is'rbf'
grid1.fit(X_train_scaled,y_train)

# grid search end
end_time = time.time()

# Evaluate results
print("Best Parameters:", grid1.best_params_)
print("Best Score:", grid1.best_score_)

# print total time for grid search completion 
total_time_secd = end_time - start_time
total_time_minu = total_time_secd/ 60
print(f" Gird search took {total_time_minu: .2f} minutes to complete.")
   
```

```python
# Make the predication on X_test
y_pred1=grid1.predict(X_test_scaled)
print(y_pred1)
```


- **5.2 SVM Model Evaluation**

```python
# Evaluate the model predication on y_test & y_pred: Evaluation criteria used:Accuracy, Recall, Precision  
acc_svm= accuracy_score(y_test,y_pred1)
rec_svm= recall_score(y_test,y_pred1)
prec_svm= precision_score(y_test,y_pred1)
print (f"The accuracy of the designed SVM model: {acc_svm* 100:.1f}%")
print (f"The recall of the designed SVM model: {rec_svm* 100:.1f}%")
print (f"The precision of the designed SVM model: {prec_svm* 100:.1f}%")
# calculate f score
fscore_svm= f1_score(y_test, y_pred1, average='binary')
print(f" The F-score for the designed SVM model: {fscore_svm *100:.1f}%")
# Print the Confusion Matrix
cm_svm = confusion_matrix(y_test, y_pred1)
print('True Positives(TP) = ', cm_svm[0,0])
print('True Negatives(TN) = ', cm_svm[1,1])
print('False Positives(FP) = ', cm_svm[0,1])
print('False Negatives(FN) = ', cm_svm[1,0])
```

- **5.3 RF  Design and Training**

```python

# define the model
classifier2 = RandomForestClassifier() 

# Hyperparameter tuning for the RF model using GridSearchCV
param_grid2 = {
    'n_estimators': [10, 20, 40],  # number of trees in the forest
    'max_depth': [ None, 5, 10, 15 ],  # Max number of levels in tree
    'min_samples_split': [2, 5, 10],  # Min number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4],  # Min number of samples required to be at a leaf node
    'criterion':['gini','entropy'],
}

# grid search start
start_time = time.time()

# Do a grid search 
grid2 = GridSearchCV(estimator= classifier2, param_grid= param_grid2, scoring = 'accuracy', cv= 5 , n_jobs=-1)

# Fit the model- # Train the RF  Classifier with X_train & y_train 
grid2.fit(X_train_scaled, y_train)

# grid search end
end_time = time.time()

# Evaluate results
print("Best Parameters:", grid2.best_params_)
print("Best Score:", grid2.best_score_)

# print total time for grid search completion 
total_time_secd = end_time - start_time
total_time_minu = total_time_secd/ 60
print(f" Gird search took {total_time_minu: .2f} minutes to complete.")
```

```python
# Make the predication on X_test
y_pred2=grid2.predict(X_test_scaled)
print(y_pred2)
```

 - **5.4 RF  Model Evaluation**


```python
# Evaluate the model predication on y_test & y_pred: Evaluation criteria used:Accuracy, Recall, Precision  
acc_rf= accuracy_score(y_test,y_pred2)
rec_rf= recall_score(y_test,y_pred2)
prec_rf= precision_score(y_test,y_pred2)
print (f"The accuracy of the designed RF model: {acc_rf* 100:.1f}%")
print (f"The recall of the designed RF model: {rec_rf* 100:.1f}%")
print (f"The precision of the designed RF model: {prec_rf* 100:.1f}%")

# calculate f score
fscore_rf = f1_score(y_test, y_pred2, average='binary')
print(f" The F-score for the designed RF model: {fscore_rf *100:.1f}%")

# Print the Confusion Matrix
cm_rf = confusion_matrix(y_test, y_pred2)
print('True Positives(TP) = ', cm_rf[0,0])
print('True Negatives(TN) = ', cm_rf[1,1])
print('False Positives(FP) = ', cm_rf[0,1])
print('False Negatives(FN) = ', cm_rf[1,0])
```

- **7.2 Key Findings**

**For Full dataset**: 
- RF has the highest performance with 99.7% accuracy, and stable performance with 99.6% recall, precision, and F1-score. - SVM achieves an overall accuracy of 97.3% and the  highest precision of 99.9%, while its recall and F1-score are slightly lower with 94.1% and 96.9
    respectively.
  
**For current Dataset**:
- The accuracy of RF improved by 0.2 % achieving 99.9% overall accuracy, while its recall and F1-score decreased to 1.5 % and 0.7% respectively compared    to the full dataset.
- The accuracy, recall, and F1-score of SVM deteriorated around 1.8%, 4%, and 2.1% respectively, yet achieved the best precision with 100%.
  
**For Voltage Dataset**:
- SVM and RF models showed a reduction of 3.3% and 4.6% respectively inthe  accuracy performance. Yet, RF still outperformed with 95.1% accuracy compared
  to 94% for SVM.
- RF deteriorated significantly to a 7% reduction in the recall, 5.7 % in F1-score, and 3.2 % in  precision.
- SVM’s performance is the most influenced with a 12.6% reduction in its recall performance. 
   
 



- **7.3 Key Takeaways**
  
- Overall, RF outperforms SVM for the three datasets regarding accuracy, recall, and F1-score, however, SVM exhibits higher precision compared to RF. The highest accuracy achieved is linked to the RF model by  utilizing the current dataset with an overall accuracy of 99.9 %.
- The precision of SVM is higher due to the algorithm foundation to minimize misclassifications (i.e., reduces false positives) close to the decision boundary, yet increasing false negatives, resulting in higher precision and lower recall.
  
**The potential reasons for the lower accuracy performance of the voltage data in both models**:
- Voltage observations demonstrate less variability than current data (i.e., lower standard deviations), which leads to not capturing the significant fluctuations that may occur under fault conditions.
- The voltage reading may not be significant enough to capture some types of faults (e.g., short circuits and overloads) that are correlated with current flow behavior.
- The statistical descriptive data show that voltage readings are strongly clustered around zero, resulting limitation of providing a real separation between classes and misclassification of a high number of false positives or false negatives that worsen the overall accuracy.

- **7.4 Future works**

-  Applying probability calibration to enhance the reliability and the significance of the model's performance. [Probability calibration](https://scikit-learn.org/1.5/modules/calibration.html). 
- Adding noise to the data to evaluate how changes in variance affect the performance(i.e., models' robustness).
